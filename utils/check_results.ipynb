```python
import pandas as pd
```

```python
import xml2csv
import dotdict
import xml.sax
```

```python
source = '../Experiment/Results/test3.xml'
```

```python
options_dict = {'separator':';', 'quotechar':'', 'xsd':None, 'validation':False, 'split':False, 'output':None, 'source':source}
options = dotdict.dotdict(options_dict)
```

```python
options.source
```

```python
attrFinder = xml2csv.AttrFinder(options.xsd, options.source, options.split)
```

```python
handler = xml2csv.CSVWriter(attrFinder, options)
```

```python
xml.sax.parse(options.source, handler)
```
